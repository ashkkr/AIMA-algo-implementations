```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0,2 * np.pi,300)
siny = np.sin(x);
cosy = np.cos(x);

plt.plot(x,siny)
plt.plot(x,cosy)
plt.xlabel("x")
plt.title("Sin and Cos Curve")
plt.axhline(0, color='black', linewidth=0.8)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

mean = 0
std = 1

x = np.linspace(-20,20,300);
y1 = norm.pdf(x,mean,1)
y2 = norm.pdf(x,mean,5)
y3 = norm.pdf(x,mean,10)


plt.plot(x,y1);
plt.plot(x,y2)
plt.plot(x,y3)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm

a = 1.0
b = 1.0

# Axis
h = np.linspace(-10,10,100)
c = np.linspace(-10,10,100)
H,C = np.meshgrid(h,c)

mean = a * H + b
std = 1

Z = norm.pdf(C,mean,std)

# Plot

fig = plt.figure()
ax = fig.add_subplot(111,projection = "3d")

ax.plot_surface(H,C,Z,cmap = "viridis")
ax.set_xlabel("H")
ax.set_ylabel("C")
ax.set_zlabel("Z")

plt.show()
```
